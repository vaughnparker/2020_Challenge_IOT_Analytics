```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import h5py
import os
```

```python
def hdf_file_to_df(hdf_file_name):
    cwd = os.getcwd()
    f = h5py.File(hdf_file_name, 'r')
    dd_keys = list(f['DYNAMIC DATA'].keys())
    
    #matrix = np.array()
    matrix = []
    
    for key in dd_keys:
        one_channel = np.array(f['DYNAMIC DATA'][key]['MEASURED'])
        matrix.append(one_channel)
        #print(len(matrix))
    
    np_matrix = np.array(matrix)
    return pd.DataFrame(np_matrix.T, columns=dd_keys)
```

```python
def find_file(year, month, day):
    file_list = os.listdir('competitionfiles/')
    return_list = []
    
    for filename in file_list:
        if (filename[8:12] == year and
            filename[12:14] == month and
            filename[14:16] == day):
            return_list.append(filename)
    
    return return_list
```

* basically, we need to find which channels are in which files
* to do this, we will get a list of all channels, then check each file for each channel

```python
def get_all_channels_list():
    filename_list = os.listdir('competitionfiles/')[:10]
    channels_list = []
    
    for i in range(len(filename_list)):
        filename = filename_list[i]
        print('Loading: File #' + str(i) + ': ' + filename)
        df = hdf_file_to_df('competitionfiles/' + filename)
        
        for channel in df.columns:
            if channel not in channels_list:
                channels_list.append(channel)
                print('added: ' + channel)
                
    return channels_list
```
